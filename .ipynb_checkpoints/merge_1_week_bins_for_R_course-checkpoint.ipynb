```python
import pandas as pd
import numpy as np
import os
import statsmodels.formula.api as sm
```

```python
week_dir = '1_week_bins/'
lac1_gpa = pd.read_csv('lac1_freshmen_gpa.csv')
lac2_gpa = pd.read_csv('lac2_freshmen_gpa.csv')
weeks = os.listdir(week_dir)    
```

```python
def getAllSubjects():
    idxs = []
    for week in weeks:
        num_week = week.split('_')[0]
        filename = os.path.join(week_dir, week)
        df = pd.read_csv(filename)
        df['uid'] = df[['subject_id', 'cohort']].apply(lambda x: '_'.join(x.values.astype(str)), axis=1)
        idxs.extend(list(df['uid']))
    return list(set(idxs))
```

```python
getAllSubjects()
```

```python
dfs = []
idxs = getAllSubjects()
for week in weeks:
    num_week = week.split('_')[0]
    df = pd.read_csv(os.path.join(week_dir, week))
    df['uid'] = df[['subject_id', 'cohort']].apply(lambda x: '_'.join(x.values.astype(str)), axis=1)
    df = df.set_index('uid')
    df = df.drop(['subject_id', 'cohort', 'preCESD_sum', 'postCESD_sum', 'frac_nights_with_data'], axis=1)
    df.columns = [str(col) + '_' + num_week for col in df.columns]
    df.reindex(idxs)
    dfs.append(df)
```

```python
all_df = pd.concat(dfs, axis=1)
```

```python
lac1_gpa['uid'] = lac1_gpa['subject_id'].apply(lambda x: str(x)+'_lac1')
```

```python
lac2_gpa['uid'] = lac2_gpa['subject_id'].apply(lambda x: str(x)+'_lac2')
```

```python
all_df
```

```python
df = all_df.merge(lac1_gpa, 
                  right_on='uid', 
                  left_index=True, how='outer').merge(lac2_gpa, 
                                                      on='uid', 
                                                      how='outer')
df['cum_gpa'] = df['cum_gpa_x'].fillna(df['cum_gpa_y'])
df['term_gpa'] = df['term_gpa_x'].fillna(df['term_gpa_y'])
df = df.set_index(['uid'])
df = df.drop(['subject_id_x', 'subject_id_y', 
              'cum_gpa_x', 'cum_gpa_y',
              'term_gpa_x', 'cum_gpa_x'], axis=1)
```

```python
df.to_csv('stats_course_lac_sleep_gpa_data.csv')
```

```python
result = sm.ols(formula="term_gpa ~ midpoint_sleep_6 + cum_gpa", data=df).fit()
print(result.summary())
```
